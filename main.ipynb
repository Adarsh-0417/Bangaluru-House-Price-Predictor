```python
def convertRange(x):
    temp = x.split('-')
    if len(temp) == 2:
        return (float(temp[0])+float(temp[1]))/2
    try: 
        return float(x)
    except:
        return None
```

```python
data['total_sqft']=data['total_sqft'].apply(convertRange)
```

```python
data.head()
```

```python
data['price-per-sqft']=data['price']*100000/data['total_sqft']
```

```python
data['location'] = data['location'].apply(lambda x: x.strip())
```

```python
location_count =data['location'].value_counts()
```

```python
location_count
```

```python
location_count_less_than_10 = location_count[location_count<=10]
location_count_less_than_10
```

```python
data['location']=data['location'].apply(lambda x : 'other' if x in location_count_less_than_10 else x)
```

Outlier detection and handling

```python
data.describe()
```

```python
(data['total_sqft']/data['bhk']).describe()
```

```python
data = data[((data['total_sqft']/data['bhk']) >= 300)]
data.describe()
```

```python
data['price-per-sqft'].describe()
```

```python
def remove_outliers_sqft(df):
    df_output = pd.DataFrame()
    for key,subdf in df.groupby('location'):
        m = np.mean(subdf['price-per-sqft'])
        st = np.std(subdf['price-per-sqft'])
        gen_df = subdf[(subdf['price-per-sqft'] > (m-st)) &  (subdf['price-per-sqft'] <= (m+st))]
        df_output = pd.concat([df_output,gen_df], ignore_index=True)
    return df_output
data = remove_outliers_sqft(data)
data.describe()
    
```
